```python
# Block 5: Import libraries and load cleaned data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set path to the cleaned CSV file
cleaned_path = r"E:\file_main\Major_1\fauna-forecast\data\cleaned\Bison_bison_cleaned.csv"

# Load the cleaned data
df = pd.read_csv(cleaned_path)

# Basic checks
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nMissing values:\n", df.isnull().sum())

# Preview
df.head(5)
```

```python
# Block 6: Total observations per year

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='year', palette='viridis')
plt.title("Number of Observations per Year")
plt.xlabel("Year")
plt.ylabel("Count of Records")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Block 7: Monthly distribution (seasonal trends)

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='month', palette='coolwarm')
plt.title("Monthly Observation Pattern")
plt.xlabel("Month")
plt.ylabel("Number of Observations")
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()
```

```python
# Block 8: Mean individualCount per year (if column exists)

if 'individualCount' in df.columns:
    yearly_counts = df.groupby('year')['individualCount'].mean()

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=yearly_counts, marker='o')
    plt.title("Mean Individual Count per Year")
    plt.xlabel("Year")
    plt.ylabel("Average Individual Count")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Column 'individualCount' not available in the dataset.")
```

```python
# Block 9: Scatter plot of geographic coordinates

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='decimalLongitude', y='decimalLatitude', hue='year', palette='Spectral', s=30, alpha=0.7)
plt.title("Geographic Distribution of Bison Observations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
